# method Overriding

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

obj = Dog()
obj.speak()
```

```python
class Vehicle:
    def move(self):
        print("Vehicle moves")

class Car(Vehicle):
    def move(self):
        print("Car moves on road")

obj = Car()
obj.move()
```

```python
class Shape:
    def draw(self):
        print("Drawing Shape")

class Circle(Shape):
    def draw(self):
        print("Drawing Circle")

obj = Circle()
obj.draw()
```

```python
class Employee:
    def work(self):
        print("Employee works")

class Developer(Employee):
    def work(self):
        print("Developer writes code")

obj = Developer()
obj.work()
```

# method overloading

```python
class Vehicle:

    def fuel_type(self):
        print("Vehicle Fuel")

    def move(self):
        print("Vehicle Moving")

    def show(self):
        self.fuel_type()
        self.move()


class Car(Vehicle):

    def fuel_type(self):
        print("Fuel : Petrol")

    def move(self):
        print("Car moves on Road")


class Bike(Vehicle):

    def fuel_type(self):
        print("Fuel : Petrol")

    def move(self):
        print("Bike moves on Road")


class Boat(Vehicle):

    def fuel_type(self):
        print("Fuel : Diesel")

    def move(self):
        print("Boat moves on Water")


class Airplane(Vehicle):

    def fuel_type(self):
        print("Fuel : Aviation Fuel")

    def move(self):
        print("Airplane flies in Sky")


class Rocket(Vehicle):

    def fuel_type(self):
        print("Fuel : Liquid Hydrogen")

    def move(self):
        print("Rocket goes to Space")


vehicles = [
    Car(),
    Bike(),
    Boat(),
    Airplane(),
    Rocket()
]

for vehicle in vehicles:
    vehicle.show()
    print("----------------")
```

```python
class Calculator:

    def add(self, a, b=0, c=0):
        print(a + b + c)

obj = Calculator()

obj.add(10)
obj.add(10, 20)
obj.add(10, 20, 30)
```

```python
class Student:

    def details(self, name, age=None, city=None):
        print(name, age, city)

obj = Student()

obj.details("Yusuf")
obj.details("Yusuf", 22)
obj.details("Yusuf", 22, "Hyderabad")
```

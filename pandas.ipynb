# HARD LEVEL

```python
#1. Salary > department average & Performance < department average
salary_avg=df.groupby("Department")["Salary"].transform("mean")

rating_avg=df.groupby("Department")["Performance_Rating"].transform("mean")

df[(df["Salary"]>salary_avg) &
   (df["Performance_Rating"]<rating_avg)]
```

```python
#2. Salary Outliers (IQR)
def outliers(group):

    q1=group["Salary"].quantile(0.25)

    q3=group["Salary"].quantile(0.75)

    iqr=q3-q1

    lower=q1-1.5*iqr

    upper=q3+1.5*iqr

    return group[(group["Salary"]<lower) |
                 (group["Salary"]>upper)]

df.groupby("Department",group_keys=False).apply(outliers)
```

```python
#3. Promotion Risk
median=df.groupby("Department")["Salary"].transform("median")

df[(df["Experience_Years"]>8)&
   (df["Performance_Rating"]>=4)&
   (df["Salary"]<median)]
```

```python
#4. Average experience > company average
company=df["Experience_Years"].mean()

dept=df.groupby("Department")["Experience_Years"].mean()

dept[dept>company]
```

```python
#5. Correlation
df[["Age","Salary",
"Experience_Years",
"Performance_Rating"]].corr()
```

```python
#6. High Risk Attrition
median=df["Salary"].median()

df[(df["Age"]<30)&
   (df["Salary"]<median)&
   (df["Performance_Rating"]<=2)]
```

```python
#7. Department Salary Rank
df["Salary_Rank"]=df.groupby("Department")["Salary"].rank(ascending=False)

df
```

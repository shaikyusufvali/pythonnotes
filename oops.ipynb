```python
class Student:
    def __init__(self,st_name,rollno):
        self.Std_name = st_name
        self.Roll_no = rollno

    def display_student(self):
        print("name:",self.Std_name)
        print("roll no:",self.Roll_no)
class collegeStudent(Student):
    def __init__(self,Std_name,Roll_no,course_name):
        super().__init__(Std_name,Roll_no)
        self.course_name = course_name
    def display_course(self):
        self.display_student()
        print("course:",self.course_name)
s = collegeStudent("yusuf",21,"Ai")

s.display_course()
        
```

```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def show_vehicle(self):
        print("Brand:", self.brand)
        print("Model:", self.model)


class Bike(Vehicle):
    def __init__(self, brand, model, mileage):
        Vehicle.__init__(self, brand, model)
        self.mileage = mileage

    def show_bike(self):
        self.show_vehicle()
        print("Mileage:", self.mileage)


b = Bike("Honda", "Shine", 60)
b.show_bike()
```

```python
class Father:
    def __init__(self, father_name):
        self.father_name = father_name

class Mother:
    def __init__(self, mother_name):
        self.mother_name = mother_name

class Child(Father, Mother):
    def __init__(self, father_name, mother_name, child_name):
        Father.__init__(self, father_name)
        Mother.__init__(self, mother_name)
        self.child_name = child_name

    def display(self):
        print("Father:", self.father_name)
        print("Mother:", self.mother_name)
        print("Child:", self.child_name)

c = Child("Ahmed", "Ayesha", "Yusuf")
c.display()
```

```python
#Multiple inheritance
class Account:
    def __init__(self,account_no, **kwargs):
        self.account_no = account_no
        super().__init__(**kwargs)
class Customer:
    def __init__(self,customer_name, **kwargs):
        self.customer_name = customer_name
        super().__init__(**kwargs)
class Bank(Account,Customer):
    def __init__(self,account_no,customer_name,bank_name):
        self.Bank_name = bank_name
        super().__init__(account_no=account_no,customer_name=customer_name)

    def display(self):
        print("account no:",self.account_no)
        print("customer_name:", self.customer_name)
        print("bank_name:",self.Bank_name)
obj = Bank(786868,"YUSUF","hdfc")
obj.display()
    
```

# 1 . Hybrid Inheritance

```python
class Company:
    def __init__(self, Company_Name, Company_City, **kwargs):
        super().__init__(**kwargs)
        self.Company_Name = Company_Name
        self.Company_City = Company_City


class HR(Company):
    def __init__(self, HR_Manager, **kwargs):
        super().__init__(**kwargs)
        self.HR_Manager = HR_Manager


class IT(Company):
    def __init__(self, Programming_Language, **kwargs):
        super().__init__(**kwargs)
        self.Programming_Language = Programming_Language


class Employee(HR, IT):
    def __init__(self, Company_Name, Company_City, HR_Manager,
                 Programming_Language, Employee_Name):

        super().__init__(
            Company_Name=Company_Name,
            Company_City=Company_City,
            HR_Manager=HR_Manager,
            Programming_Language=Programming_Language
        )

        self.Employee_Name = Employee_Name

    def display(self):
        print("Company :", self.Company_Name)
        print("City :", self.Company_City)
        print("HR Manager :", self.HR_Manager)
        print("Programming Language :", self.Programming_Language)
        print("Employee :", self.Employee_Name)


emp = Employee("TCS", "Hyderabad", "Ramesh", "Python", "Rahul")
emp.display() 
```

```python
class Hospital:
    def __init__(self, Hospital_Name, Location, **kwargs):
        super().__init__(**kwargs)
        self.Hospital_Name = Hospital_Name
        self.Location = Location


class Doctor(Hospital):
    def __init__(self, Doctor_Name, **kwargs):
        super().__init__(**kwargs)
        self.Doctor_Name = Doctor_Name


class Nurse(Hospital):
    def __init__(self, Nurse_Name, **kwargs):
        super().__init__(**kwargs)
        self.Nurse_Name = Nurse_Name


class Patient(Doctor, Nurse):
    def __init__(self, Hospital_Name, Location,
                 Doctor_Name, Nurse_Name, Patient_Name):

        super().__init__(
            Hospital_Name=Hospital_Name,
            Location=Location,
            Doctor_Name=Doctor_Name,
            Nurse_Name=Nurse_Name
        )

        self.Patient_Name = Patient_Name

    def display(self):
        print(self.Hospital_Name)
        print(self.Location)
        print(self.Doctor_Name)
        print(self.Nurse_Name)
        print(self.Patient_Name)


obj = Patient("Apollo", "Hyderabad", "Dr. Ravi", "Anitha", "Suresh")
obj.display()
```

```python
class University:
    def __init__(self, University_Name, City, **kwargs):
        super().__init__(**kwargs)
        self.University_Name = University_Name
        self.City = City


class College(University):
    def __init__(self, College_Name, **kwargs):
        super().__init__(**kwargs)
        self.College_Name = College_Name


class School(University):
    def __init__(self, School_Name, **kwargs):
        super().__init__(**kwargs)
        self.School_Name = School_Name


class Student(College, School):
    def __init__(self, University_Name, City,
                 College_Name, School_Name, Student_Name):

        super().__init__(
            University_Name=University_Name,
            City=City,
            College_Name=College_Name,
            School_Name=School_Name
        )

        self.Student_Name = Student_Name

    def display(self):
        print(self.University_Name)
        print(self.City)
        print(self.College_Name)
        print(self.School_Name)
        print(self.Student_Name)


obj = Student("JNTU", "Hyderabad", "JNTUH College", "JNTUH School", "Ali")
obj.display()
```

```python
class Bank:
    def __init__(self, Bank_Name, Branch, **kwargs):
        super().__init__(**kwargs)
        self.Bank_Name = Bank_Name
        self.Branch = Branch


class Savings(Bank):
    def __init__(self, Savings_Account_No, **kwargs):
        super().__init__(**kwargs)
        self.Savings_Account_No = Savings_Account_No


class Loan(Bank):
    def __init__(self, Loan_Amount, **kwargs):
        super().__init__(**kwargs)
        self.Loan_Amount = Loan_Amount


class Customer(Savings, Loan):
    def __init__(self, Bank_Name, Branch,
                 Savings_Account_No, Loan_Amount, Customer_Name):

        super().__init__(
            Bank_Name=Bank_Name,
            Branch=Branch,
            Savings_Account_No=Savings_Account_No,
            Loan_Amount=Loan_Amount
        )

        self.Customer_Name = Customer_Name

    def display(self):
        print(self.Bank_Name)
        print(self.Branch)
        print(self.Savings_Account_No)
        print(self.Loan_Amount)
        print(self.Customer_Name)


obj = Customer("SBI", "KPHB", 12345678, 500000, "Rahul")
obj.display()
```

# 2. Multiple Inheritance

```python
class Company:
    def __init__(self, company_name, city):
        self.company_name = company_name
        self.city = city


class Employee(Company):
    def __init__(self, company_name, city, employee_name):
        super().__init__(company_name, city)
        self.employee_name = employee_name

    def display(self):
        print("Company :", self.company_name)
        print("City :", self.city)
        print("Employee :", self.employee_name)


obj = Employee("TCS", "Hyderabad", "Rahul")
obj.display()
```

#Multiple Inheritance

```python
class HR:
    def __init__(self, hr_manager, **kwargs):
        super().__init__(**kwargs)
        self.hr_manager = hr_manager


class IT:
    def __init__(self, programming_language, **kwargs):
        super().__init__(**kwargs)
        self.programming_language = programming_language


class Employee(HR, IT):
    def __init__(self, hr_manager, programming_language, employee_name):

        super().__init__(
            hr_manager=hr_manager,
            programming_language=programming_language
        )

        self.employee_name = employee_name

    def display(self):
        print("HR Manager :", self.hr_manager)
        print("Programming Language :", self.programming_language)
        print("Employee :", self.employee_name)


obj = Employee("Ramesh", "Python", "Rahul")
obj.display()
```

# 3. Multilevel Inheritance

```python
class Company:
    def __init__(self, company_name):
        self.company_name = company_name


class Manager(Company):
    def __init__(self, company_name, manager_name):
        super().__init__(company_name)
        self.manager_name = manager_name


class Employee(Manager):
    def __init__(self, company_name, manager_name, employee_name):
        super().__init__(company_name, manager_name)
        self.employee_name = employee_name

    def display(self):
        print("Company :", self.company_name)
        print("Manager :", self.manager_name)
        print("Employee :", self.employee_name)


obj = Employee("Infosys", "Suresh", "Rahul")
obj.display()
```

# Hierarchical Inheritance

```python
class Company:
    def __init__(self, company_name):
        self.company_name = company_name


class HR(Company):
    def __init__(self, company_name, hr_manager):
        super().__init__(company_name)
        self.hr_manager = hr_manager

    def display(self):
        print("Company :", self.company_name)
        print("HR Manager :", self.hr_manager)


class IT(Company):
    def __init__(self, company_name, programming_language):
        super().__init__(company_name)
        self.programming_language = programming_language

    def display(self):
        print("Company :", self.company_name)
        print("Programming Language :", self.programming_language)


obj1 = HR("Wipro", "Ramesh")
obj2 = IT("Wipro", "Python")

obj1.display()
obj2.display()
```

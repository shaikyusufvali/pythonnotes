```python
# create array : 1d
import numpy as np
array_1d = np.array([2,3,4,5,7])
type(array_1d)
```

```python
# create Special NumPy Arrays, 1. Zeros Array
import numpy as np

a = np.zeros((2,3))
print(a)
```

```python
# create Special NumPy Arrays, 1. ones Array
a = np.ones((3,2))
print(a)
```

# 1.sum()

# 2. mean() → Average

```python
import numpy as np
a =np.array([10,20,30,40,50,60]) # addition method or % 5
print(np.mean(a))
```

# 3. max() → Largest Value

```python
import numpy as np
a =np.array([10,20,30,40,50,60]) # max value
print(np.max(a))
```

# 4. min() → Smallest Value

```python
import numpy as np
a =np.array([10,20,30,40,50,60]) # smallest value
print(np.min(a))
```

# 5. std() → Standard Deviation

```python
import numpy as np
a =np.array([10,20,30,40,50,60]) 
print(np.std(a))
```

# 6. var() → Variance

```python
import numpy as np
a =np.array([10,20,30,40,50,60]) 
print(np.var(a))
```
